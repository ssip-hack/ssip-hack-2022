```python
def data_set(a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s):
    val = (a**19)+(b**18)+(c**17)+(d**16)+(e**15)+(f**14)+(g**13)+(h**12)+(i**11)+(j**10)+(k**9)+(l**8)+(m**7)+(n**6)+(o**5)+(p**4)+(q**3)+(r**2)+s
    f= val % 100
    if f >= 0 and f <= 99:
        return float(f/100)
    if f >= 100 and f<=999:
        return float(f/1000)
    else:
        return float(f/10000)
```
