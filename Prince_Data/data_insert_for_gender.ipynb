```python
import pandas as pd
o1 = pd.read_csv('Final_main.csv')
o1
```

```python
age_list=[]
for index in range(0,50000):
    age_list.append(o1['gender'][index])
```

```python
fg=[]
for i in age_list:
    if i==1:
        fg.append('male')
    else:
        fg.append('female')
```

```python
o1['gender_data'] = fg
```
