```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('dropout.csv')
```

```python
data.columns
```

# for the Base on School

```python
def Primary_School_Prediction(data):
    # split the data for the Train and test
    x = data[['primary_Girls' , 'primary_Boys' ]]
    y = data['primary_Overall']
    x_train, x_test , y_train , y_test = train_test_split(x,y,random_state=56, test_size = 0.6 )
    
    # run the actual model for the prediction
    lr = LinearRegression()
    lr.fit(x_train , y_train)
    
    # print statements
    print("Co-efficient : --> ", lr.coef_)
    print("intercepts_ : ", lr.intercept_)
    
    # prediction
    prediction = lr.predict(x_test)
    final_outcome = (y_test - prediction)
    
    # plot the difference data for better understanding 
    #sns.distplot(final_outcome)    
    
    # metrics 
    m1 = metrics.mean_absolute_error(y_test , prediction)
    m2= metrics.mean_squared_error(y_test, prediction)
    print(f"Mean Square : {m2} \n Absolute error: {m1}" )
```

# Plot The 3D Graph

```python
from sklearn.cluster import KMeans
import plotly.express as px

pdata = data[['primary_Girls' ,'primary_Boys' , 'primary_Overall']].copy()
km = KMeans(n_clusters=5)
data['Data'] = km.fit_predict(pdata)
data['Data'] = data['Data'].astype('category')
px.scatter_3d(data , x = 'primary_Girls' , y = 'primary_Boys' , z='primary_Overall' )
```
